# Pengantar Git Version dan Github
## Git
Git adalah program yang digunakan untuk menyimpan dan mengelola versi-versi file pekerjaan. Pengeloaan file tersebut dilakukan di dalam komputer secara lokal.

Ilustrasi versi file dapat diibaratkan seperti kita membuat file dokumen (file_1.doc, file_2.doc, dan seterusnya sampai file_final.doc) dimana nomor menunjukkan versi kebaruan dari file pekerjaan. Daripada membuat dan menyimpan banyak file dengan nomor yang berbeda setiap kita melakukan perbaikan dokumen, maka program <b>Git</b> membantu dalam menyimpan <b> hanya versi perubahan </b> dalam file dokumen yang sedang dikerjakan. Dengan demikian, cukup membuat sebuah nama file.doc, kemudian setiap ada perubahan atau perbaikan, kita menyimpan informasi perubahan ke dalam program Git tersebut.

Keuntungan sederhana lainnya menggunakan program Git adalah kita bisa melihat perubahan-perubahan pada setiap versi file dan apabila diperlukan, kita <i>set back</i> ke versi terdahulu/lama dari suatu file yang sedang dikerjakan.

## Github
<a href="github.com">Github </a> adalah tempat menyimpan file-file versi pekerjaan dalam suatu repositori / tempat penyimpanan secara online. Github adalah salah satu penyedia, ada banyak penyedia ruang  penyimpanan versi pekerjaan online lainnya. Keuntungan menggunakan tempat penyimpanan secara online untuk versi-versi file pekerjaan adalah

1. dapat diakses dari berbagai perangkat secara online

2. dapat diakses oleh lebih dari satu orang untuk bekerja secara kolaboratif

## Instalasi Git dan Registrasi Akun Github

Sesi ini belum menjelaskan bagaimana proses instalasi git dan registrasi akun github. Diasumsikan pembaca bisa melakukan secara mandiri prose-proses

1. Download program <a href="https://git-scm.com/downloads"> Git </a>

2. Instalasi program Git.

3. Registrasi akun pada <a href="github.com">Github </a>


## Membuat Folder/Repositori pada <a href="github.com">Github </a>

1. Login menggunakan akun github.

2. Klik tombol tab <b>Repositories</b> >> <b> New </b>

3. Isikan nama folder atau repo yang diinginkan. Sebagai contoh, foldernya bernama <b>git-version</b>.

4. Tuliskan deskripsi singkat mengenai folder yang dibuat. Contoh: <i> Folder ini menyimpan tutorial sederhana penggunaan git dan github</i>.

5. Check list pada bagian [ ] Add a README file dan abaikan lebih dahulu pilihan-pilihan lainnya.

6. Klik tombol <b> Create Repository </b>

## Membuat Clonning Folder dari <a href="github.com">Github </a> ke dalam Folder Komputer Lokal

Ada beberapa cara membuatnya, namun disini dcontohkan satu cara saja.

1. Diasumsikan Anda saat ini sudah login dan berada pada folder github yang telah dibuat sebelumnya. Contoh url folder, https://github.com/priyandari/git-version

2. Klik tanda panah pada tombol tombol Code yang berwarna hijau.

3. Pilih atau tekan pilihan Download Zip untuk mendapatkan clone folder dengan nama git-version-main.zip

4. Copy file git-version-main.zip ke dalam suatu tempat yang Anda inginkan. Kemudian ekstraksi file zip tersebut sehingga Anda memperoleh folder dengan nama <b> git-version-main</b>

5. Misalkan folder tersebut berada pada lokasi hardisk F:/git-version-main . Kalau diperhatikan, di dalam folder tersebut minimal ada sebuah file dengan nama README.md

6. Gunakan interface command line (cmd) atau terminal untuk mengeksekusi perintah-perintah Git pada folder di atas yang akan dijelaskan pada bagian selanjutnya.


## Config dan Init

Langkah awal adalah melakukan konfigurasi terkait hak akses.

1. Buka interface command line (cmd) atau terminal untuk mengeksekusi perintah-perintah Git.

2. Pastikan terminal sudah berada di folder clonning yang dibuat sebelumnya. gunakan perintah <i> pwd </i>

3. lakukan proses konfigurasi menggunakan perintah berikut ini

> git config --global user.username "Your User Detail Name"

> git config --global user.email youremail@gmail.com

> git config --global user.username "yourusername"

4. Check git config.

> git config --list


Langkah berikutnya membuat atau konfigurasi folder f:/git-version-main agar dikenali sebagai folder untuk mengelola file pekerjaan berbasis git version.

1. Gunakan perintah ini:

> git init .

perhatikan tanda titik tersebut. Artinya menjadi folder dilokasi saat ini menjadi folder berbasis git.

> Initialized empty Git repository in F:/git-version-main/.git/

2. Jika Anda menampilkan hidden folder, maka bisa dilihat ada folder <b>.git</b>

## Mengarahkan Remote / Sinkronisasi antara Folder Lokal dan Repo di <a href="github.com">Github </a>

Langkah berikutnya ada mengatur agar folder di lokal komputer sebelumnya dapat disinkronkan dengan folder atau repositori di <a href="github.com">Github </a>

1. Buka repositori <a href="github.com">Github </a> yang telah dibuat

2. klik tombol Code berwarna hijau, lalu pilih tab HTTPS, dan klik tombol Copy untuk menyalin url alamat repositori

3. Ketik perintah ini. Perhatikan alamat url adalah copy-paste dari langkah 2.

> git remote add github https://github.com/priyandari/git-version.git

Catatan: struktur dasarnya seperti di bawah ini. Nama singkat itu akan <b>digunakan</b> saat kita akan melakukan proses upload ke folder repo

> git remote add <namasingkat> <link URL HTTPS>


4. Periksa apakah remote-remote sudah ditambahkan

> git remote -v
